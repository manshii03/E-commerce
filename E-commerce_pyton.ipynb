```python
sale['Gender'].value_counts().plot(kind='bar',color=['salmon','hotpink'])
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.ylim(100,3000)
plt.xticks(rotation=45)
plt.show()
```

```python
sale['Category'].value_counts().plot(kind='pie',autopct='%1.1f%%',shadow=True ,explode=[0,0,0,0.125])
plt.ylabel(' ')
plt.title('Category')
plt.show()
```

```python
sale['Season'].value_counts().plot(kind='bar', color='hotpink')
plt.title('seasons')
plt.xticks(rotation=45)
plt.show()
```

```python
sale['Discount Applied'].value_counts().plot(kind='pie',autopct='%1.2f%%',startangle=50, explode=(0.05,0),shadow=True)
plt.title('Discount Applied')
plt.ylabel('')
plt.show()
```

```python
sale['Item Purchased'].value_counts().plot(kind='bar')
plt.title('Item Purchased')
plt.tight_layout()
plt.show()
```

```python
sale.hist(bins=6,edgecolor='yellow',figsize=(10,15),color='green')
plt.show()
```
